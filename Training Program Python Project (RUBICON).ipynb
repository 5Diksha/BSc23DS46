```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("/content/best-selling-books.csv")
```

```python
# Check dataset shape (rows, columns)
print("\nDataset shape:", df.shape)
```

```python
# Column names
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Display few rows
print("Preview of the dataset:")
print(df.head())
```

```python
#Missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Drop rows with missing data
df.dropna(inplace=True)
```

```python
# Summary statistics for numerical columns
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Top 10 most frequent authors
top_authors = df['Author(s)'].value_counts().head(10)
print("\nTop 10 Authors:\n", top_authors)
```

```python
# Unique genres
print("\nUnique genres:")
print(df['Genre'].unique())
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Genre', order=df['Genre'].value_counts().index, palette='Set1')
plt.title("Number of Books per Genre")
plt.xlabel("Number of Books")
plt.ylabel("Genre")
plt.yticks(fontsize=5)
plt.tight_layout()
plt.show()
```

```python
# Count of books per genre
print("\nBooks per Genre:")
print(df['Genre'].value_counts())
```

```python
# Books with highest sales
print("\nTop 10 Best-Selling Books:")
print(df.sort_values(by="Approximate sales in millions", ascending=False).head(10))
```

```python
# Line Chart Top 10 Best-Selling Books

top_books = df.nlargest(10, "Approximate sales in millions")

plt.figure(figsize=(10, 5))
plt.plot(top_books["Book"],top_books["Approximate sales in millions"], marker="o", linestyle="-", color="b")
plt.xticks(rotation=90,fontsize=5)
plt.xlabel("Book")
plt.ylabel("Sales (millions)")
plt.title("Top 10 Best-Selling Books")
plt.grid()
plt.show()
```

```python
# Top 10 Authors by Number of Books
top_authors = df['Author(s)'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette='magma')
plt.title("Top 10 Authors by Number of Best-Selling Books")
plt.xlabel("Number of Books")
plt.ylabel("Author")
plt.tight_layout()
plt.show()
```

```python
# Price Distribution by Genre

df["Approximate sales in millions"] = pd.to_numeric(df["Approximate sales in millions"], errors="coerce")

sales_bins = pd.cut(df["Approximate sales in millions"], bins=[0, 20, 50, 100, 200], labels=["0-20M", "21-50M", "51-100M", "101-200M"])
sales_distribution = sales_bins.value_counts()

plt.figure(figsize=(10, 6))
plt.pie(sales_distribution, labels=sales_distribution.index, autopct="%1.1f%%", colors=plt.cm.Paired.colors)
plt.title("Distribution of Book Sales (in millions)")
plt.show()
```

```python
# Books Published Over Time
plt.figure(figsize=(12, 6))
sns.histplot(df['First published'], bins=20, kde=False, color='orange')
plt.title("Books Published Over Time")
plt.xlabel("Year Published")
plt.ylabel("Number of Books")
plt.tight_layout()
plt.show()
```

```python
# Number of Books Published Each Year
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='First published', palette='Set2')
plt.title('Number of Best-Selling Books per Year')
plt.xlabel('Year')
plt.ylabel('Number of Books')
plt.xticks(rotation=90, fontsize=5)
plt.tight_layout()
plt.show()
```
